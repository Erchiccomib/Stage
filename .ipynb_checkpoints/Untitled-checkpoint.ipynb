## <center> __Implementazione sul dataset journal.pone.0148699_S1_Text_Sepsis_SIRS_EDITED__ </center>

Il primo algoritmo che ho implementato è stato k-means, di seguito riporto il codice:


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import seaborn as sns

#Leggo il dataset dal file csv
dataset = pd.read_csv('C:\\Users\\fncba\\OneDrive\Documenti\\Stage\\Cartelle cliniche\\journal.pone.0148699_S1_Text_Sepsis_SIRS_EDITED.csv') 

print(dataset.info()) # Da qui noto tutte le informazioni dei dati, sopratutto il tipo di dato presente così da capire se pre-elaborare i dati

#Ho notato che sono tutti di tipo float64 e int64 quindi vanno bene per l'elaborazione tramite k-means

scalar = MinMaxScaler() #Utilizzo MinMaxScaler per normalizzare i dati
scalar_feauter = scalar.fit_transform(dataset) #Normalizzo i dati

pca = PCA(n_components=2) #Setto PCA per ridurre la dimensionalità dei dati così da ottenere una migliora rappresentazione, setto n_components=2 per indicare le componenti da mantenere dopo la riduzione
features = pca.fit_transform(scalar_feauter) #trasformo la dimensione dei dati

K= range(2,10) #Setto il range di cluster da formare tramite k-means

#Utilizzo la tecnica del gomito per comprendere quanti cluster formare
inertia = []
for k in K: #Inizio un ciclo in cui per ogni k in K eseguo k-means con k cluster e salvo l'inerzia (somma delle sitanze quadratiche tra ogni punto del cluster ed il centroide) per comprendere con quale k è più efficiente
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)
    labelss = kmeans.labels_

#Rappresento a video la curvatura dell'inerzia per ogni k testato su k-means
plt.figure(figsize=(10,6))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Numeri di cluster (k)')
plt.ylabel('Inertia')
plt.title('Gomito')
plt.grid(True)
plt.show()

#Da qui ho notato che tramite k=4 si hanno i risultati migliori, dunque ho eseguito k-means con k=4 così da mostrare il cluster ottenuto
kmenas = KMeans(n_clusters =4, random_state=42)
kmenas.fit(features) #Addestro il modello
labels = kmenas.labels_ #Ottengo le labels

dataset['Cluster'] = labels #Creo una nuova colonna 'Cluster' in cui indico a quale cluster è stato associato ogni elemento

#Calcolo e stampo a video il valore delle metriche di valutazione
score_silhouette = silhouette_score(features, labels)
print(f'Punteggio Silhouette: {score_silhouette}')

score_calinski = calinski_harabasz_score(features, labels)
print(f'Punteggio Calinski: {score_calinski} ')

score_davies = davies_bouldin_score(features, labels)
print(f'Punteggio Davies: {score_davies}')



#Mostro il cluster ottenuto
plt.figure(figsize=(12, 8)) #imposto la dimensione del grafico
scatter = plt.scatter(features[:, 0], features[:, 1], c=labels, cmap='viridis') #Creo un grafico a dispersione
centroids = kmenas.cluster_centers_ #Ottengo i centroidi dei cluster per mostrarli di rosso tramite delle X 
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X', label='Centroids') 
plt.title('K-means (k=4)')
plt.xlabel('Component 1 PCA')#Rappresenta quanto il valore di ogni punto della componente 1 di PCA si discosta dalla media dei valori
plt.ylabel('Component 2 PCA')
plt.legend()
plt.colorbar(scatter, label='Cluster Label')
plt.grid(True)
plt.show()

#Calcolo delle statistiche descrittive per ciascun cluster così da comprendere la differenza presente tra i cluster
cluster_summary = dataset.groupby('Cluster').mean()

# Heatmap delle medie delle caratteristiche nei diversi cluster
plt.figure(figsize=(12, 8))
sns.heatmap(cluster_summary.T, annot=True, cmap='viridis')
plt.title('Heatmap (Tramite StandardScaler)')
plt.xlabel('Cluster')
plt.ylabel('Feature')
plt.show()
```

Proseguiamo con l'implementazione dell'algoritmo di clustering agglomerativo (di tipo gerarchico), di seguito riporto il codice:

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns

#Leggo il dataset
dataset = pd.read_csv('C:\\Users\\fncba\\OneDrive\Documenti\\Stage\\Cartelle cliniche\\journal.pone.0148699_S1_Text_Sepsis_SIRS_EDITED.csv')

#Normalizzo il dataset tramite MinMaxScaler
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(dataset)

#Riduco la dimensione dei valori del dataset per ottenere sia una migliore rappresentazione che un miglior clustering
pca = PCA(n_components=2)
features = pca.fit_transform(features_scaled)

#specifico il criterio utilizzato per unire i cluster durante il processo di clustering gerarchico e salvo il collegamento in Z
Z = linkage(features, 'ward')

#Mostro a video il dendrogramma per comprendere il numero di cluster da formare (k=4)
plot.figure(figsize=(10,7))
dendrogram(Z)
plot.title('Dendrogramma')
plot.xlabel('Cluster')
plot.ylabel('Distance')
plot.show()

#Eseguo il cluster con k=4
agglomerative = AgglomerativeClustering(n_clusters=4, linkage="ward")
agglomerative.fit(features)

labels= agglomerative.labels_

dataset['Cluster'] = labels

#Mostro il clustering ottenuto
plot.figure(figsize=(12,8))
plot.scatter(features[:,0], features[:,1], c=labels, cmap='viridis')
plot.title("Agglomerative Clustering (k=4)")
plot.xlabel("Component 1 PCA")
plot.ylabel("Component 2 PCA")
plot.grid(True)
plot.show()

#Calcolo e stampo i valori delle metriche di valutazione
punteggio_silhouette = silhouette_score(features, labels)
print(f'Punteggio Silhouette: {punteggio_silhouette}')

punteggio_calinski = calinski_harabasz_score(features, labels)
print(f'Punteggio Calinski: {punteggio_calinski}')

punteggio_davies = davies_bouldin_score(features, labels)
print(f'Punteggio Davies: {punteggio_davies}')

#Stampo a video l'heatMap 
heatMap = dataset.groupby(['Cluster']).mean()

plot.figure(figsize=(12,8))
sns.heatmap(heatMap.T, annot=True, cmap='viridis')
plot.xlabel("Cluster")
plot.ylabel("Features")
plot.title("HeatMap")
plot.show()
```

###<center> __Spiegazione del codice Agglomerative__ </center>
La struttura del codice è simile al clustering con k-means, infatti si nota all'inizio la lettura del dataset, la normalizzazione mediante MinMaxScaler e la riduzione della dimensione dei valori tramite PCA. Andando avanti non si ha la tecnica del gomito ma si usa il dendrogramma per comprendere effettivamente quanti cluster formare, notando 4 cluster. Successivamente, ho eseguito il clustering con 4 cluster e ho mostrato sia i cluster ottenuti che i punteggi delle metriche di valutazione e l'heatMap per la comprensione della differenza dei cluster formati. Anche in questo caso i risultati ottenuti sono molto solidi ed indicano, soprattutto grazie alle metriche di valutazione, la produzione di un buon clustering. 

* Cluster 1, questo cluster rappresenta un gruppo di pazienti prevalentemente maschi, più anziani, con condizioni cliniche gravi e una mortalità elevata;
* Cluster 2, rappresenta un gruppo di pazienti prevalentemente femmine, di età simile al Cluster 1, con condizioni cliniche gravi e la mortalità più alta tra i cluster.
* Cluster 3, rappresenta un gruppo di pazienti più giovani, con condizioni cliniche meno gravi e una mortalità molto bassa.
* Cluster 4, rappresenta un gruppo di pazienti più giovani, prevalentemente maschi, con condizioni cliniche meno gravi e una mortalità molto bassa.

Ci tengo a specificare che anche in questo caso, prima dell'ottenimento del clustering mostrato, ho effettuato diversi testo con vari iperparametri e metriche di normalizzazione, ottendo una migliore efficienza con MinMaxScaler per la normalizzazione e gli iperparametri n_clusters=4, linkage="ward".


Proseguiamo con l'implementazione dell'algoritmo di clustering BIRCH, di seguito riporto il codice:

```python
from sklearn.cluster import Birch
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns

#Leggo il dataset
dataset = pd.read_csv('C:\\Users\\fncba\\OneDrive\Documenti\\Stage\\Cartelle cliniche\\journal.pone.0148699_S1_Text_Sepsis_SIRS_EDITED.csv')

#Normalizzo i valori tramite MinMaxScaler
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(dataset)

#Riduco la dimensione dei valori del dataset per ottenere sia una migliore rappresentazione che un miglior clustering
pca = PCA(n_components=2)
features = pca.fit_transform(features_scaled)

#Eseguo l'algoritmo BIRCH
birch = Birch(threshold=0.3, branching_factor=50, n_clusters=None) #Ho provato diversi valori per branching_factor ma non cambia nulla, mentre per threshold ho notato che impostandolo a 0.3 si ha un miglior clustering
birch.fit(features)

labels = birch.labels_

dataset['Cluster'] = labels

#Mostro i cluster ottenuti
plot.figure(figsize=(12,8))
plot.scatter(features[:,0], features[:,1], c=labels, cmap='viridis')
plot.title("BIRCH Clustering")
plot.xlabel("Component 1 PCA")
plot.ylabel("Component 2 PCA")
plot.grid(True)
plot.show()

#Calcolo e stampo i valori delle metriche di valutazione
punteggio_silhouette = silhouette_score(features, labels)
print(f'Punteggio Silhouette: {punteggio_silhouette}')

punteggio_calinski = calinski_harabasz_score(features, labels)
print(f'Punteggio Calinski: {punteggio_calinski}')

punteggio_davies = davies_bouldin_score(features, labels)
print(f'Punteggio Davies: {punteggio_davies}')

#Stampo l'heatMap 
heatMap = dataset.groupby(['Cluster']).mean()

plot.figure(figsize=(12,8))
sns.heatmap(heatMap.T, annot=True, cmap='viridis')
plot.xlabel("Cluster")
plot.ylabel("Features")
plot.title("HeatMap")
plot.show()
```

###<center> __Spiegazione del codice BIRCH__ </center>
Dopo aver letto il dataset, normalizzato i valori e ridotto la dimensione, ho direttamente eseguito l'algoritmo poichè non richiede una specifica del numero di cluster da formare. Dunque, ho testato l'algoritmo con diversi iperparametri e ho ottenuto una migliore efficienza con *threshold=0.3, branching_factor=50, n_clusters=None*.
Anche in questo caso ci possiamo ritenere soddisfatti del clustering ottenuto poichè i valori delle metriche di valutazione sono significativi e tramite l'immagine che rappresenta i cluster prodotti si può notare un buon clustering senza la presenza di sovrapposizioni di cluster o punti troppo distanti.
L'algoritmo ha formato 4 cluster, nello specifico:
* Cluster 1, rappresenta un gruppo di pazienti più giovani, con condizioni cliniche meno gravi e una mortalità molto bassa;
* Cluster 2, rappresenta un gruppo di pazienti anziani, prevalentemente maschi, con condizioni cliniche gravi e un'elevata mortalità;
* Cluster 3, rappresenta un gruppo di pazienti di mezza età, con condizioni cliniche meno gravi e una mortalità molto bassa, prevalentemente femmine;
* Cluster 4, rappresenta un gruppo di pazienti anziani, prevalentemente femmine, con condizioni cliniche gravi e un'elevata mortalità.


Proseguiamo con l'implementazione dell'algoritmo di clustering DBSCAN, di seguito riporto il codice:

```python
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import seaborn as sns

#Leggo il dataset
dataset = pd.read_csv('C:\\Users\\fncba\\OneDrive\Documenti\\Stage\\Cartelle cliniche\\journal.pone.0148699_S1_Text_Sepsis_SIRS_EDITED.csv')

#Normalizzo i valori tramite MinMaxScaler
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(dataset)

#Riduco la dimensione dei valori del dataset per ottenere sia una migliore rappresentazione che un miglior clustering
pca = PCA(n_components=2)
features = pca.fit_transform(features_scaled)

#Eseguo l'algoritmo DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5) #Ho provato, utilizzando la normalizzazione tramite MinMaxScaler, diversi epsilon ma non cambia il punteggio delle metriche
dbscan.fit(features)

labels = dbscan.labels_

dataset['Cluster'] = labels

#Stampo i valori delle metriche di valutazione
silhouette_punteggio = silhouette_score(features, labels)
print(f'Punteggio Silhouette: {silhouette_punteggio}')

calinski_punteggio = calinski_harabasz_score(features, labels)
print(f'Punteggio Calinski: {calinski_punteggio}')
    
davies_punteggio = davies_bouldin_score(features, labels)
print(f'Punteggio Davies: {davies_punteggio}')

# Visualizzazione dei risultati del clustering
plt.figure(figsize=(12, 8))
scatter = plt.scatter(features[:, 0], features[:, 1], c=labels, cmap='viridis', s=50, alpha=0.6)
plt.title('DBSCAN Clustering')
plt.xlabel('Component 1 PCA')
plt.ylabel('Component 2 PCA')
plt.colorbar(scatter, label='Cluster Label')
plt.grid(True)
plt.show()

#Visualizzazione HeatMap

heatMap = dataset.groupby('Cluster').mean()

plt.figure(figsize=(12,8))
sns.heatmap(heatMap.T, annot=True, cmap='viridis')
plt.title('Heatmap')
plt.xlabel('Cluster')
plt.ylabel('Feature')
plt.show()
```

###<center> __Spiegazione del codice DBSCAN__ </center>
Anche qui ho prima eseguito una pre-elaborazione dei dati, andandoli a normalizzare e riducendo la loro dimensione, successivamente ho eseguito l'algoritmo DBSCAN senza prima comprendere il numero di cluster da formare poichè DBSCAN non lo richiede come iperparametro. Ho testato DBSCAN con diverse configurazioni degli iperparametri senza ottenere nessun miglioramento sui risultati, dunque ho lasciato eps=0.5, min_samples=5.
Il punteggio delle metriche di valutazione sui cluster ottenuti non sono molto prestanti, infatti rispetto agli altri algoritmi di clustering, DBSCAN ha ottenuto una minore efficienza producendo solo due cluster:
* Cluster 1, rappresenta un gruppo di pazienti prevalentemente maschi, di età media, con condizioni cliniche moderate e una mortalità relativamente bassa.
* Cluster 2, rappresenta un gruppo di pazienti prevalentemente femmine, di età media, con condizioni cliniche moderate simili al Cluster 1, ma con una leggera differenza in termini di valori di alcune variabili e mortalità leggermente più alta.


Proseguiamo con l'implementazione dell'algoritmo Spectral Clustering, di seguito riporto il codice:

```python
from sklearn.cluster import SpectralClustering
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, pairwise_distances_argmin_min
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#Leggo il dataset
dataset = pd.read_csv('C:\\Users\\fncba\\OneDrive\Documenti\\Stage\\Cartelle cliniche\\journal.pone.0148699_S1_Text_Sepsis_SIRS_EDITED.csv')

#Normalizzo i valori tramite MinMaxScaler
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(dataset)

#Riduco la dimensione dei valori del dataset per ottenere sia una migliore rappresentazione che un miglior clustering
pca = PCA(n_components=2)
features = pca.fit_transform(features_scaled)

#Utilizzo la tecnica del gomito per comprendere quanti cluster formare
inertia=[]
K=range(2,10)

#Rispetto a k-means l'algoritmo spettrale non permette di ottenere direttamente l'inerzia, dunque va calcolata
for k in K:
    spectral = SpectralClustering(n_clusters=k, affinity='rbf', random_state=42)
    spectral.fit(features)
    labels = spectral.labels_
    centroids = np.array([features[labels == i].mean(axis=0) for i in range(k)])
    
    # Calcolo della "inerzia" come somma delle distanze quadrate dai centroidi
    closest, distances = pairwise_distances_argmin_min(features, centroids)
    inertia_ = np.sum(distances ** 2)
    inertia.append(inertia_)

# Visualizzazione del metodo del gomito
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, marker='o')
plt.xlabel('Numero di Cluster')
plt.ylabel('Inerzia')
plt.title('Metodo del Gomito per Determinare il Numero Ottimale di Cluster')
plt.grid(True)
plt.show()

#Eseguo l'algoritmo spettrale con k=4
spectral = SpectralClustering(n_clusters=4, affinity='rbf', gamma=35, random_state=42)
spectral.fit(features)

labels = spectral.labels_
dataset['Cluster'] = labels

#Mostro i cluster prodotti
plt.figure(figsize=(12,8))
plt.scatter(features[:,0], features[:,1], c=labels, cmap='viridis')
plt.xlabel("Component 1 PCA")
plt.ylabel("Component 2 PCA")
plt.grid(True)
plt.title("Spectral Clustering (k=4) con affinity='rbf' e gamma=35")
plt.show()

#Calcolo e stampo il valore delle metriche di valutazione
punteggio_silhouette = silhouette_score(features, labels)
print(f'Punteggio Silhouette: {punteggio_silhouette}')

punteggio_calinski = calinski_harabasz_score(features, labels)
print(f'Punteggio Calinski: {punteggio_calinski}')

punteggio_davies = davies_bouldin_score(features, labels)
print(f'Punteggio Davies: {punteggio_davies}')

#Mostro l'heatMap
heatMap = dataset.groupby(['Cluster']).mean()

plt.figure(figsize=(12,8))
sns.heatmap(heatMap.T, annot=True, cmap='viridis')
plt.xlabel("Cluster")
plt.ylabel("Features")
plt.title("HeatMap")
plt.show()
```

###<center> __Spiegazione del codice Spectral__ </center>
Una volta eseguita la preleaborazione dei dati ho utilizzato la tecnica del gomito per comprendere il numero di cluster da formare in base all'inerzia, ma siccome l'algoritmo spettrale non presenta un modo diretto per ottenere l'inerzia, ho eseguito un ciclo che calcolasse l'inerzia per ogni k appartenente al range (2,10), così da mostrare l'andamento dell'inerzia in base al numero di cluster utilizzati, scoprendo dunque una riduzione lineare dell'inerzia a partire da k=4 (rispecchiando la tecnica del gomito utilizzata per k-means). Quindi, ho eseguito l'algoritmo con k=4 e l'ho testato con diverse configurazioni degli iperparametri scoprendo una 
migliore efficienza (pur di poco) con affinity='rbf' e gamma = 35.
* Cluster 1, rappresenta un gruppo di pazienti più giovani, con condizioni cliniche meno gravi e una mortalità molto bassa;
* Cluster 2, rappresenta un gruppo di pazienti più anziani, prevalentemente femmine, con condizioni cliniche gravi e una mortalità moderata;
* Cluster 3,  rappresenta un gruppo di pazienti anziani, prevalentemente maschi, con condizioni cliniche molto gravi e una mortalità moderata;
* Cluster 4, rappresenta un gruppo di pazienti più giovani, prevalentemente maschi, con condizioni cliniche meno gravi e una mortalità molto bassa.

Anche questo algoritmo di clustering offre degli ottimi valori delle metriche di valutazione, rientrando tra i primi tre algoritmi di clustering che hanno prodotto un miglior lavoro sul dataset in questione. 


Concludiamo con l'ultimo algoritmo di clustering eseguito su questo dataset, cioè l'algoritmo Mean-Shift, di seguito il codice:

```python
from sklearn.cluster import MeanShift
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, pairwise_distances_argmin_min
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#Leggo il dataset
dataset = pd.read_csv('C:\\Users\\fncba\\OneDrive\Documenti\\Stage\\Cartelle cliniche\\journal.pone.0148699_S1_Text_Sepsis_SIRS_EDITED.csv')

#Normalizzo i valori tramite MinMaxScaler
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(dataset)

#Riduco la dimensione dei valori del dataset per ottenere sia una migliore rappresentazione che un miglior clustering
pca = PCA(n_components=2)
features = pca.fit_transform(features_scaled)

#Eseguo l'algoritmo Mean-Shift
mean = MeanShift(bin_seeding=True)
mean.fit(features)

labels = mean.labels_
dataset['Cluster'] = labels

#Mostro i cluster ottenuti
plt.figure(figsize=(12,8))
plt.scatter(features[:,0], features[:,1], c=labels, cmap='viridis')
plt.xlabel("Component 1 PCA")
plt.ylabel("Component 2 PCA")
plt.grid(True)
plt.title("Mean-Shift Clustering")
plt.show()

#Calcolo e stampo il punteggio delle metriche di valutazione
punteggio_silhouette = silhouette_score(features, labels)
print(f'Punteggio Silhouette: {punteggio_silhouette}')

punteggio_calinski = calinski_harabasz_score(features, labels)
print(f'Punteggio Calinski: {punteggio_calinski}')

punteggio_davies = davies_bouldin_score(features, labels)
print(f'Punteggio Davies: {punteggio_davies}')

#Mostro l'heatMap
heatMap = dataset.groupby(['Cluster']).mean()

plt.figure(figsize=(12,8))
sns.heatmap(heatMap.T, annot=True, cmap='viridis')
plt.xlabel("Cluster")
plt.ylabel("Features")
plt.title("HeatMap")
plt.show()
```

 Una volta eseguita la pre-elaborazione dei dati, ho direttamente eseguito l'algoritmo poichè non richiede il numero di cluster da formare. L'ho testato con diverse configurazioni e ho voluto utilizzare l’iperparametro bin_seeding=True per migliorare l’efficacia del clustering, mentre tutti gli altri iperparametri non influenzano il risultato.
Anche qui sono stati formati 4 cluster ottenendo un ottimo punteggio delle metriche di valutazione. Nello specifico:

* Cluster 2, rappresenta un gruppo di pazienti prevalentemente femmine, leggermente più anziane rispetto al Cluster 1, con condizioni cliniche meno gravi e una mortalità molto bassa;
* Cluster 3, rappresenta un gruppo di pazienti più anziani, prevalentemente maschi, con condizioni cliniche gravi e una mortalità moderata;
